```python
f = "x" # 첫번째 함수 지정
g = "2*x**2" # 두번째 함수 지정
h = "3*x**3" # 세번째 함수 지정

start = round(-10, 2) # 구간 시작점 지정
end = round(10, 2) # 구간 끝점 지정

x = start # 구간 초기값 대입
step = round(0.01, 2) # 구간 간격 지정
count = 0 # 반복 횟수

with open('x_input_data.txt', 'w') as f1, \ 
     open('f_output_data.txt', 'w') as f2, \
     open('g_output_data.txt', 'w') as f3, \
     open('h_output_data.txt', 'w') as f4: # x, f, g, h에 대한 텍스트 파일 생성하고 f1, f2, f3, f4로 나타냄

    while x <= end: # 구간 끝까지 도달할 때까지 아래의 명령 반복
        f1.write(str(x) + ' ') # f1에 x값을 문장 형태로 작성
        f2.write(str(eval(f)) + ' ') # f2에 f값을 문장 형태로 작성
        f3.write(str(eval(g)) + ' ') # f3에 g값을 문장 형태로 작성
        f4.write(str(eval(h)) + ' ') # f4에 h값을 문장 형태로 작성
        count += 1 # 반복횟수를 카운트

        if count % 5 == 0: # 만약 반복횟수가 5의 배수이면 다음 명령 실행
            f1.write('\n') # f1에서 줄바꿈
            f2.write('\n') # f2에서 줄바꿈
            f3.write('\n') # f3에서 줄바꿈
            f4.write('\n') # f4에서 줄바꿈
        x += step # x 값에 구간 간격만큼 더함
```

```python
with open('x_input_data.txt', 'r') as f1, \
     open('f_output_data.txt', 'r') as f2, \
     open('g_output_data.txt', 'r') as f3, \
     open('h_output_data.txt', 'r') as f4, \
     open('total_output_data.txt', 'w') as f5:
    # x, f, g, h에 대한 파일을 열어서 읽고, 새로 데이터를 저장할 txt 파일을 쓰기 모드로 열기
    
    while True: # 항상 아래의 명령 반복
        x_list = f1.readline().strip().split() # 읽은 x 데이터를 리스트 형태로 반환
        if not x_list: # x_list가 없을 시 다음 명령 실행
            break # 반복 중지
        
        sum_list = [0] * len(x_list) # f+g+h의 함숫값을 담을 리스트의 크기 설정
        f_list = list(map(float, f2.readline().strip().split())) # f2에서 읽어온 데이터를 f_list로 반환
        g_list = list(map(float, f3.readline().strip().split())) # f3에서 읽어온 데이터를 g_list로 반환
        h_list = list(map(float, f4.readline().strip().split())) # f4에서 읽어온 데이터를 h_list로 반환


        for i in range(len(x_list)): # x_list의 길이만큼 다음 명령 반복
            sum_list[i] = f_list[i] + g_list[i] + h_list[i] # f, g, h 리스트의 i번째 원소의 합을 sum_list의 i번째 원소로 선언
            f5.write(x_list[i] + ' ' + str(sum_list[i]) + '\n') # f5에 "x_list" + "sum_list" 값을 한 쌍씩 입력

```

```python
# Max, Min 

def getMax(numbers):
    result = -1000000
    for number in numbers:
        if result < number:
            result = number
    return result 

def getMin(numbers):
    result = 100000 
    for number in numbers:
        if result > number:
            result = number
    return result 

# get data

f = "-x**2+x**4"

start = round(-10) # 구간 시작점 지정
end = round(10) # 구간 끝점 지정

x = start # 구간 초기값 대입
step = 0.01 # 구간 간격 지정

y_list = []
x_list = []
val_box = []
while x <= (end):
    x_list.append(round(x, 2))
    y_list.append(round(eval(f), 2))
    x += step


for i in range(len(x_list)):
    val_box.append([x_list[i], y_list[i]])
    
# get Max, Min value

Max_val = getMax(y_list)
Min_val = getMin(y_list)

# Make result list

Max_result = []
Min_result = []

for i in range(len(y_list)):
    if y_list[i] == Max_val:
        Max_result.append(val_box[i])
    elif y_list[i] == Min_val:
        Min_result.append(val_box[i])
    
Max_result
Min_result
```
